# Importing libraries and defining constants

```python
# Imports all the libraries required for the notebook to run.
import pandas as pd
import numpy as np
import ast
import matplotlib.pyplot as plt
from google.colab import drive
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
%matplotlib inline
```

```python
# Mounts Google Drive.
drive.mount('/content/drive')
```

```python
# Sets the path for the data directory and the filenames.
DATA_PATH = "/content/drive/MyDrive/Data/"
PRE_PROCESSED_DATA_FILE_NAME = 'kaggle_newsdataset_preprocessed.csv'
RANDOM_STATE = 42
```

# Loading the dataset

```python
df = pd.read_csv(DATA_PATH + PRE_PROCESSED_DATA_FILE_NAME, low_memory=False)
```

```python
df.head()
```

## Converts list of strings represented as a single string to actual list of strings

```python
df.clean_text_tokens = df.clean_text_tokens.map(ast.literal_eval)
df.clean_title_tokens = df.clean_title_tokens.map(ast.literal_eval)
```

# Split data

```python
X_train, X_test, y_train, y_test = train_test_split(df[['title', 'norm_caps_in_title']], 
                                                    df['label'],
                                                   test_size = 0.2,
                                                   random_state = RANDOM_STATE,
                                                   stratify = df['label'])
```

```python
X_train.shape
```

```python
X_test.shape
```

# Encode the labels

```python
le = LabelEncoder()
```

```python
y_train_enc = le.fit_transform(y_train)
```

```python
y_test_enc = le.transform(y_test)
```

# Test the training set

Test if the training set on prediction rate of true news has less than 0.13% capital letters in the title of a news story and a fake news will have 0.13% or more.

```python
BREAK_POINT = 0.13
```

```python
def predict_news(percent_capitals, percentage_break_point):
    # Returns a classification for a news story.
    if percent_capitals >= percentage_break_point:
        return 0
    else:
        return 1
```

```python
y_train_enc_hat = X_train['norm_caps_in_title'].apply(lambda x: predict_news(x, BREAK_POINT))
```

```python
print(classification_report(y_train_enc, y_train_enc_hat, target_names=le.classes_))
```

```python
cm = confusion_matrix(y_train_enc, y_train_enc_hat)
```

```python
matrix = ConfusionMatrixDisplay(confusion_matrix = cm,
                                display_labels = le.classes_)
```

```python
fig, ax = plt.subplots(figsize = (8, 8))
matrix = matrix.plot(include_values = True, 
                     cmap = plt.cm.Blues, 
                     ax = ax, 
                     xticks_rotation = 'horizontal',
                     values_format = 'd')
matrix.ax_.set_title('Confusion Matrix for Training Set')
plt.show()
```

```python
cm_norm = confusion_matrix(y_train_enc, y_train_enc_hat, normalize='true')
matrix_norm = ConfusionMatrixDisplay(confusion_matrix = cm_norm,
                                     display_labels = le.classes_)
```

```python
fig, ax = plt.subplots(figsize = (8, 8))
matrix_norm = matrix_norm.plot(include_values = True, 
                               cmap = plt.cm.Blues, 
                               ax = ax, 
                               xticks_rotation = 'horizontal')
matrix_norm.ax_.set_title('Normalized Confusion Matrix for Training Set')
plt.show()
```

# **Summary of the Model**






Using the % of capital letters in the title of the news story provides 98% accuracy and is highly balanced in its miscalculations of about 2% each, although the amount of *true* news misclassified as *fake* is slightly lower than vice versa.


However, the point to be noted is: as a test, it is easily beaten by the changing tactics of fake news generators, in a similar way that classifying spam is always changing.


This text may overfit to this dataset.  There may be more *fake* news like the 2% that is misclassified in the wild and this test would fail.
